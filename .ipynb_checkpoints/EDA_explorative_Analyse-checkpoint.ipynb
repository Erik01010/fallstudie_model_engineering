# Explorative Datenanalyse
**Rolle:** Data Scientist — klare, nachvollziehbare EDA in deutscher Sprache.

**Quelldatei:** `274bc948-69cb-4540-a088-d9c9dbee4ed1.xlsx`
**Sheet(s):** ['Sheet1']

Dieses Notebook führt eine strukturierte EDA durch:
1. Daten einlesen (alle Sheets)
2. Übersicht (head, dtypes, fehlende Werte)
3. Univariate Analyse (Statistiken, Histogramme, Boxplots)
4. Bivariate Analyse (Korrelationen, Scatterplots)
5. Kategorielle Variablen (Häufigkeiten)
6. Zeitreihenanalyse (falls Datums-/Zeitspalten vorhanden)
7. Fazit & nächste Schritte

Die Zellen enthalten erklärende Kommentare und kleine, alltägliche Beispiele, um die Interpretation zu erleichtern.


```python
# Imports und Hilfsfunktionen
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 200)

def summarize_df(df):
    display(df.head(10))
    print('\n Datenformate (dtypes):')
    print(df.dtypes)
    print('\n Anzahl fehlender Werte pro Spalte:')
    print(df.isna().sum())
    print('\n Deskriptive Statistik (numerisch):')
    display(df.describe(include='number').T)

def plot_histograms(df, numeric_cols):
    for col in numeric_cols:
        plt.figure(figsize=(6,3))
        df[col].dropna().hist()
        plt.title(f'Histogramm: {col}')
        plt.xlabel(col)
        plt.ylabel('Anzahl')
        plt.tight_layout()
        plt.show()

def plot_boxplots(df, numeric_cols):
    for col in numeric_cols:
        plt.figure(figsize=(6,2.5))
        df.boxplot(column=col)
        plt.title(f'Boxplot: {col}')
        plt.tight_layout()
        plt.show()
```

```python
# Alle Sheets laden und kurz anzeigen
excel_path = "data.xlsx"
xls = pd.ExcelFile(excel_path)
sheets = xls.sheet_names
print('Gefundene Sheets:', sheets)

dfs = {name: pd.read_excel(excel_path, sheet_name=name) for name in sheets}
print('\nZeilen & Spalten pro Sheet:')
for name, d in dfs.items():
    print(f'- {name}: {d.shape[0]} Zeilen x {d.shape[1]} Spalten')
```

```python
# Haupt-DataFrame auswählen (erstes Sheet) und zusammenfassen
main_sheet = sheets[0]
df = dfs[main_sheet]
print(f'Hauptsheet: {main_sheet} — Form: {df.shape}')
summarize_df(df)
```

```python
# Spaltentypen erkennen
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
datetime_cols = [c for c in df.columns if np.issubdtype(df[c].dtype, np.datetime64)]
# Versuch, text-spalten auf Datumsformat zu prüfen
for c in df.select_dtypes(include=['object']).columns:
    try:
        parsed = pd.to_datetime(df[c], errors='coerce')
        if parsed.notna().sum() > (0.6 * len(df)):  # >60% parseable -> treat as datetime
            df[c] = parsed
            datetime_cols.append(c)
    except Exception:
        pass
# Kategorische: object oder wenige eindeutige Werte
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
for c in df.columns:
    if df[c].nunique() <= 20 and c not in categorical_cols and c not in numeric_cols and c not in datetime_cols:
        categorical_cols.append(c)

print('Numerische Spalten:', numeric_cols)
print('Datums-/Zeit-Spalten:', datetime_cols)
print('Kategorische Spalten (erste 20 oder weniger eindeutige Werte):', categorical_cols)
```

```python
# Univariate Analyse für numerische Spalten
if len(numeric_cols) == 0:
    print('Keine numerischen Spalten gefunden.')
else:
    print('Histogramme:')
    plot_histograms(df, numeric_cols)
    print('Boxplots:')
    plot_boxplots(df, numeric_cols)
```

```python
# Bivariate Analyse: Korrelationen und Scatter-Matrix
if len(numeric_cols) >= 2:
    corr = df[numeric_cols].corr()
    print('Korrelationsmatrix (Pearson):')
    display(corr)
    # Korrelation als einfache Heatmap
    plt.figure(figsize=(6,5))
    plt.imshow(corr.values, aspect='auto', interpolation='nearest')
    plt.colorbar()
    plt.xticks(range(len(corr)), corr.columns, rotation=90)
    plt.yticks(range(len(corr)), corr.columns)
    plt.title('Korrelationsmatrix (Heatmap)')
    plt.tight_layout()
    plt.show()
    # Scatter matrix (kleine Ansicht)
    print('Scatter-Matrix (kleine Auswahl, falls viele Spalten existieren):')
    to_plot = numeric_cols if len(numeric_cols) <= 6 else numeric_cols[:6]
    scatter_matrix(df[to_plot].dropna(), alpha=0.6, diagonal='hist', figsize=(8,8))
    plt.tight_layout()
    plt.show()
else:
    print('Nicht genug numerische Spalten für Korrelations- oder Scatter-Analysen.')
```

```python
# Analyse kategorialer Variablen (Häufigkeiten)
if len(categorical_cols) == 0:
    print('Keine kategorialen Spalten erkannt.')
else:
    for c in categorical_cols:
        print('\nTop-Werte für', c)
        display(df[c].value_counts(dropna=False).head(10))
        # Balkendiagramm (falls Anzahl Kategorien überschaubar)
        if df[c].nunique() <= 12:
            plt.figure(figsize=(6,3))
            df[c].value_counts(dropna=False).plot(kind='bar')
            plt.title(f'Häufigkeiten: {c}')
            plt.tight_layout()
            plt.show()
```

```python
# Zeitreihenanalyse falls Datums-/Zeitspalten vorhanden
if len(datetime_cols) == 0:
    print('Keine Datums-/Zeitspalten gefunden.')
else:
    for c in datetime_cols:
        print('\nUntersuchung der Zeitspalte:', c)
        ser = df[c].dropna()
        print('Zeitraum:', ser.min(), 'bis', ser.max())
        # Falls es numerische Messwerte gibt, aggregiere eines der numerischen Spalten über die Zeit
        if len(numeric_cols) > 0:
            sample_num = numeric_cols[0]
            ts = df.set_index(c)[sample_num].resample('D').mean() if np.issubdtype(df[c].dtype, np.datetime64) else None
            if ts is not None:
                plt.figure(figsize=(10,3))
                ts.plot()
                plt.title(f'Tagesmittel von {sample_num} über Zeit (Index: {c})')
                plt.ylabel(sample_num)
                plt.tight_layout()
                plt.show()
```

```python
# Fehlende Werte - Muster analysieren
mv = df.isna().sum()
print('Spalten mit fehlenden Werten (absteigend):')
display(mv[mv>0].sort_values(ascending=False))
# Anteil fehlender Werte
print('\nAnteil fehlender Werte pro Spalte:')
display((df.isna().mean()*100).round(2).sort_values(ascending=False).head(20))
```

## Fazit & nächste Schritte (Data Scientist)
- Identifiziere und dokumentiere Spalten mit hohem Anteil fehlender Werte (z. B. >30%) — ggf. Imputation oder Entfernen.
- Prüfe Ausreißer in numerischen Variablen (Boxplots) und entscheide, ob robustere Maße (Median) oder Transformationen nötig sind.
- Für Zeitreihen: auf konsistente Zeitauflösung resamplen (Tages-, Stundenmittel) und saisonale Muster prüfen.
- Für kategorielle Features: seltene Kategorien zusammenfassen (z. B. 'Sonstige').
- Wenn Zielvariable bekannt: EDA gezielt auf Ziel-Feature richten (Distribution, Klassenbalance).
- Nächster technischer Schritt: Feature-Engineering, fehlende Werte strategisch behandeln, dann erste Baseline-Modelle bauen.

**Hinweis:** Das Notebook ist so strukturiert, dass jede Codezelle unabhängig lauffähig ist — ideal zum schrittweisen Erkunden.
